```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
df=pd.read_excel('main_data.xlsx')
```

```python
df.rename(columns = {'SBET  a\n(m2/g)':'SA','Vtoal b\n(cm3/g)':'TPV','Vmicro c\n(cm3/g)':'MPV','C02 uptake amount':'CO2'}, inplace = True)
```

```python
df.describe()
```

```python
print(df['C'].isna().sum())
print(df['N'].isna().sum())
print(df['H'].isna().sum())
print(df['O *'].isna().sum())
```

```python
import miceforest as mf
```

```python
kds = mf.ImputationKernel(
  df,
  save_all_iterations=True,
  random_state=100
)

# Run the MICE algorithm for 2 iterations
kds.mice(iterations=100)

# Return the completed dataset.
df_imputed = kds.complete_data()
```

```python
sum=df_imputed['C']+df_imputed["H"]+df_imputed['N']+df_imputed['O *']
```
